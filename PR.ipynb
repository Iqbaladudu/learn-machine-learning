```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df_titanic = pd.read_csv('./data/titanic.csv')
df_titanic
```

```python
new_columns ={
    'Survived': 'SURVIVED',
    'Pclass': 'PCLASS',
    'Name': 'NAME',
    'Sex': 'SEX',
    'Age': 'AGE',
    'Siblings/Spouses Aboard': 'SIBSA',
    'Parents/Children Aboard': 'PARCA',
    'Fare': 'FARE'
}
```

```python
df_titanic = df_titanic.rename(columns=new_columns)
df_titanic
```

```python
sns.boxplot(data=df_titanic, x='SURVIVED', y='AGE')
plt.xlabel('Selamat')
plt.ylabel('Umur')
plt.show()
```

```python
sns.boxplot(data=df_titanic, x='SURVIVED', y='SEX')
plt.xlabel('Selamat')
plt.ylabel('Gender')
plt.show()
```

```python
sns.displot(df_titanic['SEX'], bins=20)
plt.show()
```

```python
sns.displot(df_titanic['AGE'], bins=20)
plt.show()
```

```python
sns.displot(df_titanic['SIBSA'], bins=20)
plt.show()
```

```python
sns.displot(df_titanic['PARCA'], bins=20)
plt.show()
```

```python
df_titanic.plot(kind='scatter', x='PARCA', y='SIBSA')
plt.show()
```

```python
sns.pairplot(data=df_titanic, hue='SEX')
plt.show()
```
