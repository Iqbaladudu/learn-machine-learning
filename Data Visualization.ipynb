```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df_iris = pd.read_csv('./data/iris.data', names=['sepal_length', 
                                                     'sepal_width',
                                                     'petal_length',
                                                     'petal_width',
                                                     'species'])
df_iris.head() #add .head() to show 5 first line
```

```python
print('Total rows: ', df_iris.shape[0])
print('Total columns: ', df_iris.shape[1])
```

```python
df_iris['species'].unique()
```

```python
#mencari data yang kosong
df_iris.isnull().any()
```

```python
df_iris.describe() #menampilkan deskripsi
```

# Bar Chart

```python
df_iris['species'].hist()
plt.show()
```

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

## Distplot

```python
sns.distplot(df_iris['sepal_length'], bins=20)
plt.show()
```

## Box plot

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Leng in cm')
plt.show()
```

```python
sns.boxplot(data=df_iris, x='species', y='petal_length')
plt.xlabel('Species')
plt.ylabel('Length in cm')
plt.show()
```

## Pairplot

```python
sns.pairplot(data=df_iris, hue='species')
plt.show()
```

## Heatmap

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
```
